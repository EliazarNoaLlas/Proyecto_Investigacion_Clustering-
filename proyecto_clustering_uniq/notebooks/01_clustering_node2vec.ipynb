```python
# Imputar valores nulos con la moda para variables categóricas
Postulantes['IDIOMA'].fillna(Postulantes['IDIOMA'].mode()[0], inplace=True)
Postulantes['LEE'].fillna(Postulantes['LEE'].mode()[0], inplace=True)
Postulantes['ESCRIBE'].fillna(Postulantes['ESCRIBE'].mode()[0], inplace=True)
Postulantes['HABLA'].fillna(Postulantes['HABLA'].mode()[0], inplace=True)
Postulantes['TIENE_DISCAPACIDAD'].fillna(Postulantes['TIENE_DISCAPACIDAD'].mode()[0], inplace=True)
Postulantes['TIPO_COMUNIDAD'].fillna(Postulantes['TIPO_COMUNIDAD'].mode()[0], inplace=True)
Postulantes['COMUNIDAD'].fillna(Postulantes['COMUNIDAD'].mode()[0], inplace=True)
```

#### MATRIZ DE CORRELACION UTILIZANDO LA CORRELACION DE PEARSON PARA CADA GRAFO



```python
# Crear un diccionario para almacenar las matrices de correlación por año
matrices_correlacion_por_año = {}

# Crear un diccionario para almacenar los DataFrames de similitud por año
dataframes_similitud_por_año = {}

# Iterar sobre cada DataFrame correspondiente a un año específico
for año, df in datasets_por_anio.items():
    # Obtener solo las columnas numéricas (sin UUID)
    numeric_columns = df.drop(columns=["UUID"])
    
    # Calcular la matriz de correlación de Pearson entre las filas (postulantes)
    correlation_matrix = numeric_columns.T.corr(method="pearson")
    
    # Almacenar la matriz de correlación en el diccionario utilizando el año como clave
    matrices_correlacion_por_año[año] = correlation_matrix
    
    # Obtener solo la parte triangular superior de la matriz de correlación
    upper_triangle = np.triu(correlation_matrix, k=1)
    
    # Crear un DataFrame para la matriz de similitud
    column_names = df["UUID"].tolist()
    similarity_df = pd.DataFrame(upper_triangle, columns=column_names, index=column_names)
    
    # Almacenar el DataFrame de similitud en el diccionario utilizando el año como clave
    dataframes_similitud_por_año[año] = similarity_df

# Ahora, en 'dataframes_similitud_por_año' tienes un diccionario donde cada clave es un año de postulación único
# y el valor correspondiente es el DataFrame de similitud para ese año.
```

# creamos los grafos apartir de la matriz de correlacion 


```python
import networkx as nx

# Supongamos que tienes un diccionario llamado 'dataframes_correlacion_por_año' donde cada clave es un año y el valor es el DataFrame de correlación correspondiente

# Crear un diccionario para almacenar los grafos por año
grafos_por_año = {}

# Iterar sobre cada año en el diccionario de DataFrames de correlación
for año, matriz_correlacion in matrices_correlacion_por_año.items():
    # Crear una instancia del grafo para el año actual
    G = nx.Graph()
    
    # Obtener los nombres de los postulantes (nodos)
    postulantes = matriz_correlacion.columns.tolist()

    # Agregar nodos al grafo
    G.add_nodes_from(postulantes[:])

    th = 0.80

    # Crear un conjunto para rastrear las aristas ya agregadas
    added_edges = set()

    # Agregar aristas al grafo con los valores de similitud sin repeticiones
    for i in range(len(postulantes[:])):
        for j in range(i + 1, len(postulantes[:])):
            # Filtro con el Threshold
            if abs(matriz_correlacion.iloc[i, j]) >= th:
                # Ordenar nodos para asegurarse de que estén en el mismo orden
                nodes = tuple(sorted([postulantes[i], postulantes[j]]))
                if nodes not in added_edges:
                    dist = 1 - matriz_correlacion.iloc[i, j]
                    G.add_edge(*nodes, weight=dist)
                    added_edges.add(nodes)
    
    # Almacenar el grafo en el diccionario utilizando el año como clave
    grafos_por_año[año] = G

# Ahora, en 'grafos_por_año' tienes un diccionario donde cada clave es un año de postulación único
# y el valor correspondiente es el grafo construido para ese año.
```

```python

# Año del grafo que deseas visualizar
año_deseado = 2021

# Obtener el grafo del año deseado
G = grafos_por_año[año_deseado]

# Crear una disposición de nodos para la visualización
pos = nx.spring_layout(G, seed=42)

# Obtener los pesos de las aristas para ajustar el ancho de las aristas en la visualización
edge_weights = [data['weight'] for _, _, data in G.edges(data=True)]

# Crear el grafo utilizando la disposición de nodos y los pesos de las aristas
nx.draw(G, pos, with_labels=True, node_size=100, node_color='skyblue', font_size=2, font_color='black',
        width=edge_weights, edge_color='gray', font_weight='bold')

# Ajustar el tamaño del gráfico y mostrar el grafo
plt.title(f"Grafo de Similitud entre Postulantes - Año {año_deseado}")
plt.show()
```

```python
import networkx as nx

# Crear un grafo vacío para acumular los grafos por año
grafo_acumulativo = nx.Graph()

# Iterar sobre los grafos por año
for año, grafo in grafos_por_año.items():
    # Agregar los nodos y aristas del grafo actual al grafo acumulativo
    grafo_acumulativo.add_nodes_from(grafo.nodes(data=True))
    grafo_acumulativo.add_edges_from(grafo.edges(data=True))

    # Verificar si hay nodos duplicados entre el grafo actual y el acumulativo
    nodos_acumulativo = set(grafo_acumulativo.nodes())
    nodos_actual = set(grafo.nodes())
    nodos_duplicados = nodos_acumulativo.intersection(nodos_actual)

    # Agregar aristas adicionales entre nodos duplicados
    for nodo in nodos_duplicados:
        # Obtener los vecinos del nodo en el grafo actual
        vecinos_actual = list(grafo.neighbors(nodo))
        # Obtener los vecinos del nodo en el grafo acumulativo
        vecinos_acumulativo = list(grafo_acumulativo.neighbors(nodo))
        # Agregar aristas entre los vecinos del nodo en el grafo acumulativo y el grafo actual
        for vecino in vecinos_actual:
            if vecino not in vecinos_acumulativo:
                grafo_acumulativo.add_edge(nodo, vecino)

# Visualizar el grafo acumulativo
# Puedes usar el mismo código de visualización que antes
```

Este DataFrame es útil para realizar análisis, visualizaciones y aplicar algoritmos de aprendizaje automático en el espacio de embedding. Puedes ver las primeras filas de emb_df con el método .head() para obtener una vista previa de los embeddings generados para algunos nodos específicos.


```python
from sklearn.cluster import KMeans

# Especificar el número de clusters que deseas obtener
num_clusters = 3  # Puedes ajustar este valor según tus necesidades

# Inicializar el modelo de K-Means
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

# Ajustar el modelo a la matriz de embeddings
kmeans.fit(emb_df)

# Obtener las etiquetas de cluster asignadas a cada nodo
cluster_labels = kmeans.labels_

# Agregar las etiquetas de cluster al DataFrame original
emb_df['cluster_label'] = cluster_labels

# Mostrar el resultado
print(emb_df[['cluster_label']])
```
