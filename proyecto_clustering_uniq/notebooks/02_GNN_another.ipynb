```python
import plotly.express as px
import pandas as pd

# Crear un DataFrame con los porcentajes de valores nulos por columna
null_percentages = (Postulantes.isnull().sum() / len(Postulantes)) * 100
null_df = pd.DataFrame(
    {
        "Columna": null_percentages.index,
        "Porcentaje de Valores Nulos": null_percentages.values,
    }
)

# Gr√°fico de barras para el porcentaje de valores nulos por columna
bar_fig = px.bar(
    null_df,
    x="Columna",
    y="Porcentaje de Valores Nulos",
    title="Porcentaje de Valores Nulos por Columna",
    labels={"Porcentaje de Valores Nulos": "Porcentaje"},
    template="plotly",
    color_discrete_sequence=["#1F77B4"],
)
bar_fig.update_layout(xaxis_title="Columnas", yaxis_title="Porcentaje")
bar_fig.update_xaxes(tickangle=-45, tickfont=dict(size=10))

# Gr√°fico de ubicaci√≥n de valores faltantes
missing_data_fig = px.imshow(
    Postulantes.isnull(),
    labels=dict(x="Columnas", y="Filas"),
    title="Ubicaci√≥n de Valores Faltantes",
    color_continuous_scale="Plasma",
    template="plotly",
    height=400,
)
missing_data_fig.update_layout(xaxis_title="Columnas", yaxis_title="Filas")

# Mostrar los gr√°ficos
bar_fig.show()
missing_data_fig.show()
```

# creamos la mtris de corelaciones



```python
# Crear un diccionario para almacenar las matrices de correlaci√≥n por a√±o
matrices_correlacion_por_a√±o = {}

# Crear un diccionario para almacenar los DataFrames de similitud por a√±o
dataframes_similitud_por_a√±o = {}

# Iterar sobre cada DataFrame correspondiente a un a√±o espec√≠fico
for a√±o, df in datasets_por_anio.items():
    # Obtener solo las columnas num√©ricas (sin UUID)
    numeric_columns = df.drop(columns=["UUID"])
    
    # Calcular la matriz de correlaci√≥n de Pearson entre las filas (postulantes)
    correlation_matrix = numeric_columns.T.corr(method="pearson")
    
    # Almacenar la matriz de correlaci√≥n en el diccionario utilizando el a√±o como clave
    matrices_correlacion_por_a√±o[a√±o] = correlation_matrix
    
    # Obtener solo la parte triangular superior de la matriz de correlaci√≥n
    upper_triangle = np.triu(correlation_matrix, k=1)
    
    # Crear un DataFrame para la matriz de similitud
    column_names = df["UUID"].tolist()
    similarity_df = pd.DataFrame(upper_triangle, columns=column_names, index=column_names)
    
    # Almacenar el DataFrame de similitud en el diccionario utilizando el a√±o como clave
    dataframes_similitud_por_a√±o[a√±o] = similarity_df

# Ahora, en 'dataframes_similitud_por_a√±o' tienes un diccionario donde cada clave es un a√±o de postulaci√≥n √∫nico
# y el valor correspondiente es el DataFrame de similitud para ese a√±o.
```

# creacion de la matriz de adyacenciacreamos el grafo a aprti rde una unica matris de relacion

```python
import networkx as nx

# Supongamos que tienes un diccionario llamado 'dataframes_correlacion_por_a√±o' donde cada clave es un a√±o y el valor es el DataFrame de correlaci√≥n correspondiente

# Crear un diccionario para almacenar los grafos por a√±o
grafos_por_a√±o = {}

# Iterar sobre cada a√±o en el diccionario de DataFrames de correlaci√≥n
for a√±o, matriz_correlacion in matrices_correlacion_por_a√±o.items():
    # Crear una instancia del grafo para el a√±o actual
    G = nx.Graph()
    
    # Obtener los nombres de los postulantes (nodos)
    postulantes = matriz_correlacion.columns.tolist()

    # Agregar nodos al grafo
    G.add_nodes_from(postulantes[:])

    th = 0.80

    # Crear un conjunto para rastrear las aristas ya agregadas
    added_edges = set()

    # Agregar aristas al grafo con los valores de similitud sin repeticiones
    for i in range(len(postulantes[:])):
        for j in range(i + 1, len(postulantes[:])):
            # Filtro con el Threshold
            if abs(matriz_correlacion.iloc[i, j]) >= th:
                # Ordenar nodos para asegurarse de que est√©n en el mismo orden
                nodes = tuple(sorted([postulantes[i], postulantes[j]]))
                if nodes not in added_edges:
                    dist = 1 - matriz_correlacion.iloc[i, j]
                    G.add_edge(*nodes, weight=dist)
                    added_edges.add(nodes)
    
    # Almacenar el grafo en el diccionario utilizando el a√±o como clave
    grafos_por_a√±o[a√±o] = G

# Ahora, en 'grafos_por_a√±o' tienes un diccionario donde cada clave es un a√±o de postulaci√≥n √∫nico
# y el valor correspondiente es el grafo construido para ese a√±o.
```

```python
# Asumiendo que tienes un diccionario llamado grafos_por_a√±o donde cada clave es un a√±o y el valor es el grafo correspondiente

# A√±o del grafo que deseas visualizar
a√±o_deseado = 2021

# Obtener el grafo del a√±o deseado
G = grafos_por_a√±o[a√±o_deseado]

# Crear una disposici√≥n de nodos para la visualizaci√≥n
pos = nx.spring_layout(G, seed=42)

# Obtener los pesos de las aristas para ajustar el ancho de las aristas en la visualizaci√≥n
edge_weights = [data['weight'] for _, _, data in G.edges(data=True)]

# Crear el grafo utilizando la disposici√≥n de nodos y los pesos de las aristas
nx.draw(G, pos, with_labels=True, node_size=100, node_color='skyblue', font_size=2, font_color='black',
        width=edge_weights, edge_color='gray', font_weight='bold')

# Ajustar el tama√±o del gr√°fico y mostrar el grafo
plt.title(f"Grafo de Similitud entre Postulantes - A√±o {a√±o_deseado}")
plt.show()
```

```python
import networkx as nx

# Crear un grafo vac√≠o para acumular los grafos por a√±o
grafo_acumulativo = nx.Graph()

# Iterar sobre los grafos por a√±o
for a√±o, grafo in grafos_por_a√±o.items():
    # Agregar los nodos y aristas del grafo actual al grafo acumulativo
    grafo_acumulativo.add_nodes_from(grafo.nodes(data=True))
    grafo_acumulativo.add_edges_from(grafo.edges(data=True))

    # Verificar si hay nodos duplicados entre el grafo actual y el acumulativo
    nodos_acumulativo = set(grafo_acumulativo.nodes())
    nodos_actual = set(grafo.nodes())
    nodos_duplicados = nodos_acumulativo.intersection(nodos_actual)

    # Agregar aristas adicionales entre nodos duplicados
    for nodo in nodos_duplicados:
        # Obtener los vecinos del nodo en el grafo actual
        vecinos_actual = list(grafo.neighbors(nodo))
        # Obtener los vecinos del nodo en el grafo acumulativo
        vecinos_acumulativo = list(grafo_acumulativo.neighbors(nodo))
        # Agregar aristas entre los vecinos del nodo en el grafo acumulativo y el grafo actual
        for vecino in vecinos_actual:
            if vecino not in vecinos_acumulativo:
                grafo_acumulativo.add_edge(nodo, vecino)

# Visualizar el grafo acumulativo
# Puedes usar el mismo c√≥digo de visualizaci√≥n que antes
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class GraphNeuralNetwork(nn.Module):
    def __init__(self, in_features, out_features, num_heads=1, dropout=0.6, alpha=0.2):
        super(GraphNeuralNetwork, self).__init__()
        self.in_features = in_features
        self.out_features = out_features
        self.num_heads = num_heads
        self.dropout = dropout
        self.alpha = alpha

        self.W = nn.Parameter(torch.zeros(size=(in_features, out_features)))
        nn.init.xavier_uniform_(self.W.data, gain=1.414)

        self.a = nn.Parameter(torch.zeros(size=(2*out_features, 1)))
        nn.init.xavier_uniform_(self.a.data, gain=1.414)

        self.leakyrelu = nn.LeakyReLU(self.alpha)
        self.dropout_layer = nn.Dropout(self.dropout)

    def forward(self, input, adj):
        h = torch.mm(input, self.W)
        N = h.size()[0]

        # Multiplicaci√≥n de matrices para calcular el valor de atenci√≥n
        a_input = torch.cat([h.repeat(1, N).view(N*N, -1), h.repeat(N, 1)], dim=1).view(N, -1, 2*self.out_features)
        e = self.leakyrelu(torch.matmul(a_input, self.a).squeeze(2))

        # M√°scara para ignorar nodos no conectados
        zero_vec = -9e15*torch.ones_like(e)
        attention = torch.where(adj > 0, e, zero_vec)
        attention = F.softmax(attention, dim=1)
        attention = self.dropout_layer(attention)

        # Agregaci√≥n de informaci√≥n de nodos vecinos ponderada por atenci√≥n
        h_prime = torch.matmul(attention, h)

        return h_prime

# Utilizar tus matrices de caracter√≠sticas y de adyacencia
matriz_caracteristicas_gnn = torch.randn(matriz_caracteristicas.shape[0], matriz_caracteristicas.shape[1])  # Tu matriz de caracter√≠sticas
adj_matrix_gnn = torch.tensor(adj_matrix, dtype=torch.float)  # Tu matriz de adyacencia

# Construir y entrenar el modelo GNN-AM
num_features = matriz_caracteristicas_gnn.shape[1]
model_gnn = GraphNeuralNetwork(num_features, 16)
optimizer_gnn = torch.optim.Adam(model_gnn.parameters(), lr=0.01)

# Entrenamiento
epochs = 50
for epoch in range(epochs):
    optimizer_gnn.zero_grad()
    output_gnn = model_gnn(matriz_caracteristicas_gnn, adj_matrix_gnn)
    loss_gnn = torch.mean(output_gnn)
    loss_gnn.backward()
    optimizer_gnn.step()

# Obtener los vectores de caracter√≠sticas para todos los nodos
with torch.no_grad():
    output_gnn = model_gnn(matriz_caracteristicas_gnn, adj_matrix_gnn)
    print("Vectores de caracter√≠sticas GNN-AM:", output_gnn)

# Verificar el n√∫mero de vectores obtenidos
num_vectors_gnn = output_gnn.shape[0]
print("N√∫mero de vectores de caracter√≠sticas GNN-AM obtenidos:", num_vectors_gnn)
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import pandas as pd

# Reducir la dimensionalidad con PCA
pca = PCA(n_components=2)  # Especifica el n√∫mero de componentes principales deseados
emb_df_pca = pca.fit_transform(emb_df)

# Aplicar K-Means clustering
num_clusters = 3  # N√∫mero de clusters deseado
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(emb_df_pca)

# Obtener las etiquetas de cluster asignadas a cada muestra
cluster_labels = kmeans.labels_

# Agregar las etiquetas de cluster al DataFrame original
emb_df['cluster_label'] = cluster_labels

# Mostrar el resultado
print(emb_df[['cluster_label']])
```
